### Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Link to data source: https://www.kaggle.com/aungpyaeap/supermarket-sales

**Context**

The growth of supermarkets in most populated cities are increasing and market competitions are also high. The dataset is one of the historical sales of supermarket company which has recorded in 3 different branches for 3 months data.

**Data Dictionary**

1. ***Invoice id:*** Computer generated sales slip invoice identification number

2. ***Branch:*** Branch of supercenter (3 branches are available identified by A, B and C).

3. ***City:*** Location of supercenters

4. ***Customer type:*** Type of customers, recorded by Members for customers using member card and Normal for without member card.

5. ***Gender:*** Gender type of customer

6. ***Product line:*** General item categorization groups - Electronic accessories, Fashion accessories, Food and beverages, Health and beauty, Home and lifestyle, Sports and travel

7. ***Unit price:*** Price of each product in USD

8. ***Quantity:*** Number of products purchased by customer

9. ***Tax:*** 5% tax fee for customer buying

10. ***Total:*** Total price including tax

11. ***Date:*** Date of purchase (Record available from January 2019 to March 2019)

12. ***Time:*** Purchase time (10am to 9pm)

13. ***Payment:*** Payment used by customer for purchase (3 methods are available â€“ Cash, Credit card and Ewallet)

14. ***COGS:*** Cost of goods sold

15. ***Gross margin percentage:*** Gross margin percentage

16. ***Gross income:*** Gross income

17. ***Rating:*** Customer stratification rating on their overall shopping experience (On a scale of 1 to 10)

### Task 2: Univariate Analysis

**Question 1:** What does the distribution of customer ratings looks like? Is it skewed?

```python
sns.distplot(df['Rating'])
plt.axvline(x=np.mean(df['Rating']), c='red',ls='--', label='mean')
plt.axvline(x=np.percentile(df['Rating'],25), c='green',ls='--',label='25-75th percentile')
plt.axvline(x=np.percentile(df['Rating'],75), c='green',ls='--')
plt.legend()
# From the distribtuion there does not seem to be any skewing and it looks uniformed on the graph.
```

```python
df.hist(figsize=(8,12))
# Unit price, Quantity, and Rating are uniform in their distribtuion.
# Tax, Total, Cogs, Gross Income are right skewed, as Total, Cogs, Gross Income are highly correlated.
# Gross Margin Percentage is a constant value which is the reason it does not have much distribution.
```

### Task 4: Dealing With Duplicate Rows and Missing Values

```python
df.duplicated()
# True means that is a duplicate
```

```python
df.duplicated().sum()
# there are 3 duplicates
```

```python
df[df.duplicated()==True]
```

```python
df.drop_duplicates(inplace=True)
# used to drop the duplicate values
```

```python
df.duplicated().sum()
# showing that they were dropped
```

```python
df.isna().sum()
# total number of missing values per column
```

```python
df.isna().sum()/len(df)
# looking at the ratio of missing values
# Customer type have 7.9% missing
```

```python
sns.heatmap(df.isnull(),cbar=False);
# visualizing the missing values
# cbar removes the bar on the right 
```

```python
df.fillna(df.mean(),inplace=True);
# inplace= True is a permanent change
# Filling the data with mean of each column for the missing data
# this only works for the numeric columns
```

```python
sns.heatmap(df.isnull(),cbar=False);
# showing the changes for the numeric columns because there cannot be a mean of categorical data
# Remaining data is for only the categorical 
```

```python
df.mode().iloc[0]
# This is Mode for each column
```

```python
df.fillna(df.mode().iloc[0], inplace=True)
# Each categorical column missing values are replaced with the Mode of each column
```

```python
df.fillna(0, inplace=True)
# Showing how you can fill the missing date with 0
# This is showing in Unit Price column which is a numeric column because it does not work for categorcial data.
# Did not place the command inplace=True so it did not safe did this change
```

```python
sns.heatmap(df.isnull(),cbar=False);
# No more missing values in columns 
```
